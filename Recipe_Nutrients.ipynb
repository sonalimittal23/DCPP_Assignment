```python
import pandas as pd
import numpy as np
from pandas import DataFrame
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup 
import re
import csv
import requests
import time
```

```python
recipe_url = pd.read_csv('D:/ISB/TERM 1/Data Collection/GROUP PROJECT/URL.csv')
```

```python
recipe_urls = recipe_url['URL']
```

```python
total_values=len(recipe_urls)
total_values
```

```python
calorie_list =[]
carbs_list=[]
Fat_list=[]
Protein_list=[]
Sodium_list=[]
Potassium_list=[]
Cholestrol_list=[]
VitaminA_list=[]
VitaminC_list=[]
Calcium_list=[]
Iron_list=[]
```

```python
count = 0
for url in recipe_urls:
    
  time.sleep(0.1)
    # count = count+1
  count = count+1
  print(f'{count} of {total_values}')
  page = requests.get(url)
  soup = BeautifulSoup(page.text,'lxml')
  urldish = soup.find('div',class_='jss64')
  urllink = urldish.find('a').get('href')
  urllinkfull = 'https://www.myfitnesspal.com' + urllink
  page1 = requests.get(urllinkfull)
  soup1 = BeautifulSoup(page1.text,'lxml')
  calorie = soup1.find('span',class_='title-cgZqW').text
  carbs = soup1.find_all('span',class_='title-1P2uF')[0].text
  Fat =  soup1.find_all('span',class_='title-1P2uF')[1].text
  Protein = soup1.find_all('span',class_='title-1P2uF')[2].text
  Nutri = soup1.find_all('div',class_='jss95')
  Sodium = Nutri[9].find('span').text
  Potassium = Nutri[10].find('span').text
  Cholestrol = Nutri[11].find('span').text
  VitaminA = Nutri[12].find('span').text
  VitaminC = Nutri[13].find('span').text
  Calcium = Nutri[14].find('span').text
  Iron = Nutri[15].find('span').text
  

  calorie_list.append(calorie)
  carbs_list.append(carbs)
  Fat_list.append(Fat)
  Protein_list.append(Protein) 
  Sodium_list.append(Sodium)
  Potassium_list.append(Potassium)
  Cholestrol_list.append(Cholestrol)
  VitaminA_list.append(VitaminA)
  VitaminC_list.append(VitaminC)
  Calcium_list.append(Calcium)
  Iron_list.append(Iron)

table = pd.DataFrame({'Calorie':calorie_list,'Carbs':carbs_list,'Fat':Fat_list,'Protein':Protein_list,'Sodium':Sodium_list,'Potassium':Potassium_list,'Cholestrol':Cholestrol_list,'VitaminA':VitaminA_list,'VitaminC':VitaminC_list,'Calcium':Calcium_list,'Iron':Iron_list})

```

```python
table['Fat'] = table['Fat'].replace({'g':'','--':0},regex=True)
table['Carbs']=table['Carbs'].replace({'g':'','--':0}, regex=True)
table['Protein']=table['Protein'].replace({'g':'','--':0}, regex=True)
table['Sodium']=table['Sodium'].replace({'mg':'','--':0}, regex=True)
table['Potassium']=table['Potassium'].replace({'mg':'','--':0}, regex=True)
table['Cholestrol']=table['Cholestrol'].replace({'mg':'','--':0}, regex=True)
table['VitaminA']=table['VitaminA'].replace({'%':'','--':0}, regex=True)
table['VitaminC']=table['VitaminC'].replace({'%':'','--':0}, regex=True)
table['Calcium']=table['Calcium'].replace({'%':'','--':0}, regex=True)
table['Iron']=table['Iron'].replace({'%':'','--':0}, regex=True)
```

```python
table.to_csv('calorie_file.csv',encoding='utf-8',na_rep='Unknown')
```
